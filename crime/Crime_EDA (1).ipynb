## Visual Exploration
Visual exploration was performed to identify patterns, trends, and disparities in crime occurrences across different regions using graphical representations.

```python
#. Crime Count by Area 
plt.figure(figsize=(12,6))
sns.countplot(y="AREA NAME", data=data, order=data["AREA NAME"].value_counts().index)
plt.title("Crime Count by Area")
plt.xlabel("Number of Crimes")
plt.ylabel("Area Name")
plt.show()
```

## üîπ Interpretation

*The plot highlights areas with high and low crime frequencies

* Certain areas consistently report significantly higher crime counts

* Crime distribution is uneven across regions, indicating spatial concentration

* Helps identify high-risk zones for further analysis

## üîπ Why This Visualization Matters

* Assists in geographical crime analysis

* Useful for resource allocation and policy planning

* Supports feature importance selection for machine learning models

* Provides intuitive insights before statistical modeling

## üîπ Key Insight

Area-wise crime distribution is not uniform, making location a critical feature for predictive modeling.

```python
# Crime Trend by Year
Year_count = data["Year"].value_counts().sort_index()

sns.lineplot(x=Year_count.index, y=Year_count.values)
plt.title("Crime Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.show()
```

## üìå Insight:
This line plot shows how crime occurrences change over different years. It helps identify long-term trends such as increasing, decreasing, or stable crime patterns over year.

```python
# Crime Count by Month
plt.figure(figsize=(10,5))
month_counts = data["Month"].value_counts().sort_index()
sns.barplot(x=month_counts.index , y=month_counts.values)
plt.title("Crime Count by Month")
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.show()
```

## üìå Insight:
This bar plot shows how crime occurrences vary across different months. It helps identify months with higher or lower crime rates, indicating possible seasonal trends in criminal activity.

```python
#Crime Distribution by Hour
plt.figure(figsize=(10,5))
sns.histplot(data["TIME OCC"], bins=24, kde=True)
plt.title("Crime Distribution by Hour of Day")
plt.xlabel("Time (24-hr format)")
plt.ylabel("Frequency")
plt.show()
```

## üìå Insight:
This histogram shows how crimes are distributed across different hours of the day. It helps identify peak hours when crimes are more frequent, which can be useful for time-based analysis and prevention planning.

```python
data1 = data.copy()
# Keep only valid LAT/LON values
data = data[(data["LAT"] >= 33) & (data["LAT"] <= 35)]
data = data[(data["LON"] <= -117) & (data["LON"] >= -119)]
```

```python
# Crime Hotspots Map
plt.figure(figsize=(4,4))
sns.scatterplot(x="LON",y="LAT",data=data,s=3,alpha=0.5)
plt.title("Crime Hotspots (LAT vs LON)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

## üìå Insight:
This scatter plot visualizes the geographical distribution of crime incidents using latitude and longitude. Dense clusters of points indicate crime hotspot areas where incidents occur more frequently.
